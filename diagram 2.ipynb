### LINE PLOT

```python
sns.lineplot(color="Green",x='rating',y= 'advance' , data=df)
plt.title("rating vs advance")
plt.show()
```

### PAIR PLOT

```python
 sns.pairplot(data=df)
plt.show()
```

### BAR PLOT

```python
plt.figure(figsize=(8,4))
sns.barplot('learning','complaints', data=df)
plt.title("learning & complaints")
plt.show()
```

### FACTOR PLOT

```python
sns.set_style('ticks')
sns.factorplot('rating', 'complaints', hue='raises',data=df)
plt.show()
```
