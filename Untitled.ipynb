### KDE Plot

### Line Plot

```python
sns.lineplot(color="Green",x='age',y= 'charges' , data=data)
plt.title("age vs charges")
plt.show()
```

```python
sns.set_style('ticks')
sns.factorplot("smoker", 'age', hue="sex",data=data)
plt.show()
```

```python
age_bmi = data.groupby(['age'])['children'].sum()
```

```python
age_bmi
```

```python
sns.set_style('ticks')
age_bmi.plot.pie()
plt.legend()
plt.show()
```

```python
sns.set_style('dark')
plt.figure(figsize=(4,3.5))
sns.violinplot("sex","age", hue="region", data=data)
plt.show()
             
```

### Pair Plot
